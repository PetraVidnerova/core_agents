```python
from networkx import eigenvector_centrality, degree_centrality
import pandas as pd

from graph import Graph
```

```python
g = Graph("data/twitter/twitter_combined.txt")
```

```python
G = g.to_nx_graph()
centrality = eigenvector_centrality(G)

centrality2 = degree_centrality(G)
```

```python
df = pd.DataFrame([centrality, centrality2], index=["eigenvector_centrality", "degree_centrality"]).T
```

```python
import seaborn as sns
```

```python
sns.stripplot(data=df)
```
