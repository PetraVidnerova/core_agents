```python
import pandas as pd
```

```python
df = pd.read_csv("TippingModel_twitter_results_eigenvalue.csv").rename(columns={"Unnamed: 0": "T"})
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, axs = plt.subplots(ncols=3, figsize=(18,4))
for i, beta in enumerate(df["beta"].unique()):
    sns.lineplot(data=df[df["beta"]==beta], 
        x="T", 
        y="Active", 
        hue="seed", ax=axs[i])
    axs[i].set(xlim=(0,30))
    #axs[i].set(ylim=(0,81000))
    axs[i].set_title(f"beta = {beta}")
```
